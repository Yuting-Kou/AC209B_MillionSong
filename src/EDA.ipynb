```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

data source and description: [here](https://think.cs.vt.edu/corgis/csv/music/music.html)

```python
df = pd.read_csv(r"../data/music.csv")
```

```python
df.describe()
```

```python
num_cols = df.describe().columns
```

```python
num_cols=num_cols.drop('release.id')
```

```python
plt.figure(figsize=(18,18))
plt.subplots_adjust(wspace =0.2, hspace =0.4)
for i,col in enumerate(num_cols):
    plt.subplot(5,5,i+1)
    if col=='duration':
        plt.hist(np.log(df[col]))
        plt.title('log ('+col+')',fontsize=13)
    else:
        plt.hist(df[col])
        plt.title(col,fontsize=13)
```
